```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv('/content/Salary_Data.csv')
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
print("b=",regressor.coef_)
print("a =",regressor.intercept_)
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
# regressor.predict(x_train)  because it is predicted salaries for x_train
plt.title("Simple Linear Regression on Training Data")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
import joblib

# Save the model
joblib.dump(regressor, 'linear_regression_model.pkl')
```
